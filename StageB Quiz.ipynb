```python
import numpy as np
import pandas as pd 
import sklearn.metrics as metrics 
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'

energy_data = pd.read_csv(url)

energy_data.head(6)
```

```python
x = energy_data['T2']
y = energy_data['T6']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3, random_state=1)

x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
```

```python
#fit the sample dataset
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
pred_value = model.predict(x_test)
model.fit(x_test, y_test)
```

```python
#R2 = r2_score(y_test, pred_value)

R2 = model.score(x_test, y_test)

R2.round(2)
```

```python
energy_data.drop('date', axis=1, inplace=True)

energy_data.drop('lights', axis=1, inplace=True)

energy_data.head()
```

```python
scaler = MinMaxScaler()

scaled_df = pd.DataFrame(scaler.fit_transform(energy_data), columns=energy_data.columns, index=energy_data.index)

scaled_df
```

```python
features_df = scaled_df.drop(columns = ['Appliances'])
target_df = scaled_df.Appliances
```

```python
target_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target_df, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train) 
```

```python
predicted_values = linear_model.predict(x_test)
predicted_values
```

```python
predicted_values = linear_model.predict(x_test)
linear_model.score(x_test, y_test)
```

```python
mae = mean_absolute_error(y_test, predicted_values)
mae.round(2)
```

```python
rss = np.sum(np.square(y_test - predicted_values))
rss.round(2)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
rmse.round(3) 
```

```python
r2 = r2_score(y_test, predicted_values)
r2.round(2)
```

```python
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
linear_model_weights
```
